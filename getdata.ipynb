```python
# data period
def getDataPeriod(n):
    firstPeriod=['2021-09-30','2021-12-31','2022-03-31','2022-06-30']
    allPeriod=[]

    for i in range(n):
        tempPeriod=[]
        for j in range(4):
            thisSeason=datetime.datetime.strptime(firstPeriod[j], "%Y-%m-%d")
            nYearBefore=(thisSeason-relativedelta(years=n-i-1)).strftime("%Y-%m-%d")
            tempPeriod.append(nYearBefore)
        allPeriod.append(tempPeriod)
    return allPeriod
```
