```python
import requests
import pandas as pd
```

```python
def getRevenue(start_date,data_id):
    url = "https://api.finmindtrade.com/api/v4/data"
    parameter = {
        "dataset": "TaiwanStockMonthRevenue",
        "data_id": data_id,
        "start_date": start_date,
        "token": "", # 參考登入，獲取金鑰
    }
    data = requests.get(url, params=parameter)
    data = data.json()
    data = pd.DataFrame(data['data'])
    return data
```

```python
revenueData=getRevenue('1996-01-01','2330')
```

```python
revenueData
```

```python
revenueTable=revenueData.pivot_table(index='date',values='revenue')
revenueTable
```
