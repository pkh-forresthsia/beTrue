```python
import sqlite3
import pandas as pd
import datetime
from dateutil.relativedelta import relativedelta
```

```python
revenueData=sqlite3.connect('revenueData')
def getRevenueDf(dateString):
    df=pd.read_sql('select * from revenue'+dateString.replace('-','s'),revenueData)
    return df
def getNYearRevenue(dateString,n):
    tempDf=pd.DataFrame([])
    for i in range(n*12):
        startDate=datetime.datetime.strptime(dateString, "%Y-%m-%d")
        thisMonth=(startDate-relativedelta(months=i)).strftime("%Y-%m-%d")
        thisMonthDf=getRevenueDf(thisMonth)
        tempDf=pd.concat([tempDf,thisMonthDf],join='outer')
    return tempDf
```

```python
nYearRevenue1=getNYearRevenue('2022-10-01',25)
```

```python
df=nYearRevenue1.pivot_table(values='revenue', index='date', columns='stock_id')
# getNYearRevenue('2022-11-01',5)
```
